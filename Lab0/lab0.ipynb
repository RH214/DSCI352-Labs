```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import statsmodels.api as sm
import seaborn as sns
```

```python
# Create a dataframe with columns named 'x' and 'y' filled with random numbers
df = pd.DataFrame({"x": np.random.normal(size=50), "y": np.random.normal(size=50)})

# Create an ordinary least squares regression on the data and show summary statistics.
# This formula regresses column 'y' on 'x'
ro = sm.OLS.from_formula("y ~ x", df)
res = ro.fit()
print(res.summary())

# Plot the data
sns.regplot('x', 'y', df)
plt.show()
```

```python
# Load in advertising data
advertising = pd.read_csv('Advertising.csv')
advertising.info()
advertising.describe()
```

```python
# Linear Regression: Sales against TV
tv = sm.OLS.from_formula("sales~TV", data=advertising)
tv_m = tv.fit()
print(tv_m.summary())

# p-value of the slope: 0.000
# 95% confidence interval of the slope: 0.042 to 0.053
# r-squared=0.612

# Plot the data
sns.regplot("TV","sales",advertising)
```

In this model, Y is sales in thousands of units, and X is the TV busget in thousands of dollars. On average, when TV budgets increase by \\$1000, the unit of sales increases by 47.5. The p-value is nearly 0, so we have convincing evidence that more TV budgets increase sales. We are 95\% confident that the true increase in sales associated with $1000 more budgets is between 42 and 53. The simple regression model can explain 61.2\% of the variation of sales from the expected value.
